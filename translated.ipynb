```python
M = np.float32([[1, 0, 25],	[0, 1, 50]])
```

```python
# load the image and display it to our screen
image = cv.imread("opencv.png")
cv2_imshow(image)
# shift the image 20 pixels to the right and 50 pixels up
arr_shift = np.float32([[1, 0, 20], [0, 1,-50]])
shifted = cv.warpAffine(image, arr_shift, (image.shape[1], image.shape[0]))
cv.imwrite("shifted.png",image)
cv2_imshow(shifted)
```

```python
# load the image and display it to our screen
image = cv.imread("opencv.png")
cv2_imshow(image)
# shift the image 20 pixels to the right and 50 pixels up
arr_shift = np.float32([[1, 0, -40], [0, 1,60]])
shifted = cv.warpAffine(image, arr_shift, (image.shape[1], image.shape[0]))
cv.imwrite("shifted.png",image)
cv2_imshow(shifted)
```

```python
########### Rotate ##################
(h, w) = image.shape[:2]
(cX, cY) = (w // 2, h // 2)
# rotate our image by 45 degrees around the center of the image
M = cv.getRotationMatrix2D((cX, cY), 45, 1.0)
rotated = cv.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
cv.imwrite("Rotate45.png", rotated)
# rotate our image by -90 degrees around the image
M = cv.getRotationMatrix2D((cX, cY), -90, 1.0)
rotated = cv.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
cv.imwrite("Rotated90.png", rotated)
```

```python
# rotate our image around an arbitrary point rather than the center
M = cv.getRotationMatrix2D((10, 10), 45, 1.0)
rotated = cv.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
cv.imwrite("Rotated3.png", rotated)

```

```python
######### Resize############
dim=(100,30)
# perform the resizing
resized = cv.resize(image, dim, interpolation=cv.INTER_AREA)
print(resized.shape)
cv2_imshow(resized)
```

```python
#############Resize ##################
# let's resize our image to be 150 pixels wide, but in order to
# prevent our resized image from being skewed/distorted, we must
# first calculate the ratio of the *new* width to the *old* width
print(image.shape)
r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))
# perform the actual resizing of the image
resized = cv.resize(image, dim, interpolation=cv.INTER_AREA)
cv2_imshow( image)
cv2_imshow( resized)
```
