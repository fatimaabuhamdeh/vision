```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
# Load the images
circle_image = cv2.imread('/content/circle.jpg')
triangle_image = cv2.imread('/content/triangle.jpg')
square_image = cv2.imread('/content/square.jpg')
```

```python
# Function to increase the amount of blue in the images
def increase_blue(image):
    # Split the image into its RGB channels
    b, g, r = cv2.split(image)

    # Increase the blue channel values
    b = np.clip(b + 50, 0, 255)  # You can adjust the value 50 according to your needs

    # Merge the channels back
    modified_image = cv2.merge((b, g, r))

    return modified_image
```

```python
# Apply the function to each image
modified_circle = increase_blue(circle_image)
modified_triangle = increase_blue(triangle_image)
modified_square = increase_blue(square_image)
```

```python
# Display the original and modified images
cv2_imshow(circle_image)
cv2_imshow(modified_circle)

cv2_imshow(triangle_image)
cv2_imshow(modified_triangle)

cv2_imshow(square_image)
cv2_imshow(modified_square)
```
