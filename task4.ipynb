```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Load the image
original_image = cv2.imread('/content/circle.jpg')  # Replace 'your_image.jpg' with the actual file path
```

```python
if original_image is not None:
    cv2_imshow(original_image)

    # Apply Gaussian blur filter
    blurred_image = cv2.GaussianBlur(original_image, (5, 5), 0)  # Adjust the kernel size (5, 5) as needed

    cv2_imshow(blurred_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Error: Image not loaded.")
```
