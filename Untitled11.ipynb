```python
!pip install opencv-contrib-python
import cv2 as cv
import numpy as np
import imutils
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
img = cv.imread('image.jpg', cv.IMREAD_GRAYSCALE)
cv2_imshow(img)

```

```python
(x,thr_binary) = cv.threshold(img,127,255,cv.THRESH_BINARY)
cv2_imshow(thr_binary)
```

```python
(t,thresh2)= cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
cv2_imshow(thresh2)
```

```python
(t,thresh3)= cv.threshold(img,127,255,cv.THRESH_TRUNC)
cv2_imshow(thresh3)
```

```python
(t,thresh4)= cv.threshold(img,127,255,cv.THRESH_TOZERO)
cv2_imshow(thresh4)
```

```python
(t,thresh5)= cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
cv2_imshow(thresh5)
```

```python
th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,
cv.THRESH_BINARY,11,2)
cv2_imshow(th2)
```

```python
################# THRESH_OTSU#################
img = cv.imread('image.jpg', cv.IMREAD_GRAYSCALE)
(T, threshInv) = cv.threshold(img, 0, 150,  cv.THRESH_BINARY | cv.THRESH_OTSU)
cv2_imshow(threshInv)

```

```python
image = cv.imread("image.jpg")
cv2_imshow(image)
kernelSizes = [(3, 3),(9, 9), (15, 15)]
# loop over the kernel sizes
for (kX, kY) in kernelSizes:
  # apply an "average" blur to the image using the current kernel size
  blurred = cv.blur(image, (kX, kY))
  cv2_imshow(blurred)
```
