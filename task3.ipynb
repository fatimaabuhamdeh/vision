```python
import cv2
import ipywidgets as widgets
from IPython.display import display
from google.colab.patches import cv2_imshow
```

```python
# Load the image
original_image = cv2.imread('/content/triangle.jpg')
```

```python
def flip_image(image, direction):
    if direction == 'h':
        flipped_image = cv2.flip(image, 1)  # Horizontal flip
    elif direction == 'v':
        flipped_image = cv2.flip(image, 0)  # Vertical flip
    else:
        flipped_image = image

    return flipped_image

if original_image is not None:
    cv2_imshow(original_image)

    # Create a dropdown widget for user input
    dropdown = widgets.Dropdown(options=['', 'h', 'v'], value='', description='Choose flip:')
    display(dropdown)

    def on_change(change):
        if change['type'] == 'change' and change['name'] == 'value':
            direction = change['new']

            # Check the user input
            if direction == 'h':  # Horizontal flip
                flipped_image = flip_image(original_image, 'h')
                cv2_imshow(flipped_image)
            elif direction == 'v':  # Vertical flip
                flipped_image = flip_image(original_image, 'v')
                cv2_imshow(flipped_image)

    dropdown.observe(on_change)
else:
    print("Error: Image not loaded.")
```
