```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Load the image
original_image = cv2.imread('/content/circle.jpg')  # Replace 'your_image.jpg' with the actual file path
```

```python
if original_image is not None:
    cv2_imshow(original_image)

    # Wait for user input
    key = cv2.waitKey(0) & 0xFF

    # Check the key pressed
    if key == ord('h'):  # Horizontal flip
        flipped_image = cv2.flip(original_image, 1)
        cv2_imshow(flipped_image)
    elif key == ord('v'):  # Vertical flip
        flipped_image = cv2.flip(original_image, 0)
        cv2_imshow(flipped_image)

    cv2.destroyAllWindows()
else:
    print("Error: Image not loaded.")
```
