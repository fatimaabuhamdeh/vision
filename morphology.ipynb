```python
# import the necessary packages
!pip install opencv-contrib-python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
image = cv.imread("logoj.png")

gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
```

```python
#################  Erosion #############################
kernel = np.ones((3,2),np.uint8)
cv2_imshow( gray)
erosion = cv.erode(gray,kernel,iterations = 1)
cv2_imshow( erosion)
```

```python
##################  Erosion #############################
# load the image, convert it to grayscale, and display it to our
# screen
cv2_imshow( image)
# apply a series of erosions
for i in range(0, 3):
	eroded = cv.erode(gray, (10,3),iterations=i + 1)
	cv2_imshow( eroded)
```

```python
##################   Dilation ###########################
dilation = cv.dilate(gray,kernel,iterations = 1)
cv2_imshow(dilation)
#############################################
cv2_imshow(image)
# apply a series of dilations
#none: means 3*3 kernal
for i in range(0, 3):
	dilated = cv.dilate(gray, None, iterations=i + 1)
	cv2_imshow(dilated)
```

```python
############### simple example open ##################
image = cv.imread("openj.jpg")
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
kernel = np.ones((5,5),np.uint8)
opening = cv.morphologyEx(gray, cv.MORPH_OPEN, kernel)
cv2_imshow( gray)
cv2_imshow( opening)
```

```python
################### open ###############################
cv2_imshow(image)
kernelSizes = [(3, 3), (5, 5), (7, 7)]
# loop over the kernels sizes
for kernelSize in kernelSizes:
	# construct a rectangular kernel from the current size and then
	# apply an "opening" operation
	kernel = cv.getStructuringElement(cv.MORPH_RECT, kernelSize)
	opening = cv.morphologyEx(gray, cv.MORPH_OPEN, kernel)
	cv2_imshow( opening)
```

```python
#############close####################
image = cv.imread("closej.jpg")
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
kernel = np.ones((5,5),np.uint8)
closing = cv.morphologyEx(gray, cv.MORPH_CLOSE, kernel)
cv2_imshow( gray)
cv2_imshow( closing)

```

```python
################CLOSE ###############################
cv2_imshow(image)
kernelSizes = [(3, 3), (5, 5), (7, 7)]
# loop over the kernels sizes
for kernelSize in kernelSizes:
	# construct a rectangular kernel from the current size and then
	# apply an "opening" operation
	kernel = cv.getStructuringElement(cv.MORPH_RECT, kernelSize)
	closing = cv.morphologyEx(gray, cv.MORPH_CLOSE, kernel)
	cv2_imshow( closing)
```

```python
############ Morphlogical Gradient  simple Example#############
image = cv.imread("logoj.png")
gradient = cv.morphologyEx(image, cv.MORPH_GRADIENT, kernel)
cv2_imshow(gradient)
```

```python
################Morphological gradient ###############################
image = cv.imread("logoj.png")

gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv2_imshow(image)
kernelSizes = (3, 3)
kernel = cv.getStructuringElement(cv.MORPH_RECT, kernelSize)
gradient = cv.morphologyEx(gray, cv.MORPH_GRADIENT, kernel)
cv2_imshow( gradient)
```

```python
#################### change anchor##################
image = cv.imread("closej.jpg")
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
kernel = cv.getStructuringElement(cv.MORPH_RECT, (4, 6), anchor=(2, 3))
closing = cv.morphologyEx(gray, cv.MORPH_CLOSE, kernel)
cv2_imshow( gray)
cv2_imshow( closing)


```
